1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carrega o DataFrame com o separador CORRETO (;)
sinasc = pd.read_csv('sinasc_RO_2019.csv', sep=';')

# --- Solução da Pergunta 1: Boxplot da Idade da Mãe (IDADEMAE) ---

# 2. Configura a figura
plt.figure(figsize=(10, 6))

# 3. Cria o Boxplot
# A variável é IDADEMAE
sns.boxplot(
    x=sinasc['IDADEMAE'],
    color='skyblue'
)

# 4. Configurações de Rótulos
plt.title('Boxplot da Idade da Mãe (SINASC - RO 2019)', fontsize=16)
plt.xlabel('Idade da Mãe (Anos)', fontsize=12)

# Adiciona linha vertical para a mediana
mediana_idade = sinasc['IDADEMAE'].median()
plt.axvline(
    mediana_idade, 
    color='r', 
    linestyle='--', 
    linewidth=1.5, 
    label=f'Mediana ({mediana_idade:.0f} anos)'
)
plt.legend()

plt.show()
```

2) Faça um boxplot do peso do bebê

```python
# 2. Configura a figura
plt.figure(figsize=(12, 6))

# 3. Cria o Boxplot
# A variável é PESO
sns.boxplot(
    x=sinasc['PESO'],
    color='lightcoral'
)

# 4. Configurações de Rótulos e Contextualização
plt.title('Boxplot do Peso do Bebê ao Nascer (SINASC - RO 2019)', fontsize=16)
plt.xlabel('Peso do Bebê (Gramas)', fontsize=12)

# Adiciona linha vertical para a mediana
mediana_peso = sinasc['PESO'].median()
plt.axvline(
    mediana_peso, 
    color='darkgreen', 
    linestyle='--', 
    linewidth=1.5, 
    label=f'Mediana ({mediana_peso:.0f}g)'
)
plt.legend()

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Configura a figura
plt.figure(figsize=(12, 6))

#  Cria o Histograma
# 'kde=True' adiciona a linha de densidade para suavizar a distribuição
sns.histplot(
    data=sinasc,
    x='IDADEMAE',
    bins=20,          # Define 20 barras (faixas etárias) para uma distribuição detalhada
    kde=True,
    color='purple'
)

#  Configurações de Rótulos e Contextualização
plt.title('Histograma da Idade da Mãe (SINASC - RO 2019)', fontsize=16)
plt.xlabel('Idade da Mãe (Anos)', fontsize=12)
plt.ylabel('Frequência (Contagem)', fontsize=12)

# Adiciona linha vertical para a média
media_idade = sinasc['IDADEMAE'].mean()
plt.axvline(
    media_idade, 
    color='r', 
    linestyle='-', 
    linewidth=1.5, 
    label=f'Média ({media_idade:.1f} anos)'
)
plt.legend()

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
#Configura a figura
plt.figure(figsize=(12, 6))

# 3. Cria o Histograma
# 'kde=True' adiciona a linha de densidade
sns.histplot(
    data=sinasc,
    x='PESO',
    bins=30,          # Mais bins (30) para detalhar melhor a distribuição do peso
    kde=True,
    color='darkblue'
)

# 4. Configurações de Rótulos e Contextualização
plt.title('Histograma do Peso do Bebê ao Nascer (SINASC - RO 2019)', fontsize=16)
plt.xlabel('Peso do Bebê (Gramas)', fontsize=12)
plt.ylabel('Frequência (Contagem)', fontsize=12)

# Adiciona linha vertical para a média de peso
media_peso = sinasc['PESO'].mean()
plt.axvline(
    media_peso, 
    color='r', 
    linestyle='-', 
    linewidth=1.5, 
    label=f'Média ({media_peso:.0f}g)'
)
plt.legend()

plt.show()
```
